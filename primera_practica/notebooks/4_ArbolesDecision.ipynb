# Importamos datos

```python
import pandas as pd
df = pd.read_csv('../content/tiktok.data')
df.sample(5)
```

# Seleccionamos las variables independientes

```python
x = df.drop(columns='tiktok')
x.sample(5)
```

# Seleccionamos la variable dependiente

```python
y = df.tiktok
y.sample(5)
```

# Entrenamos al modelo

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x,y)
```

# Graficamos el árbol de decisión

```python
from sklearn.tree import plot_tree
plot_tree(decision_tree = model, feature_names=x.columns, filled = True)
```

# Modificamos la profundidad del árbol (por overfitting)

```python
model = DecisionTreeClassifier(max_depth=3)
model.fit(x,y)
plot_tree(decision_tree = model, feature_names=x.columns, filled = True)
```

# Graficamos el árbol para que se visualice mejor

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,9))
plot_tree(decision_tree = model, feature_names=x.columns, filled = True,
          fontsize=13)
```

# Verificamos "a mano" la probabilidad que una persona cualquiera utilice tiktok

```python
persona = x.sample()
persona
```

```python
3/5
```

# Realizamos lo mismo pero utilizando al modelo

```python
model.predict_proba(persona)
```

# Visualizamos todas las predicciones del modelo para el conjunto de datos de entrenamiento

```python
y_pred = model.predict(x)
y_pred
```

# Comparamos las predicciones del modelo contra los valores reales del conjunto de entrenamiento

```python
df['pred'] = y_pred
df.sample(10)[['tiktok','pred']]
```

# Vemos en cuántos casos coincide la predicción con el valor real

```python
df['tiktok'] == df['pred']
```

```python
(df['tiktok'] == df['pred']).sum()
```

# Entonces, ¿qué tan bueno es el modelo?

```python
(df['tiktok'] == df['pred']).sum() / y.shape
```

```python
(df['tiktok'] == df['pred']).mean()
```

Esto quiere decir que nuestro modelo es bueno un 86,67%

```python
model.score(x, y)
```
